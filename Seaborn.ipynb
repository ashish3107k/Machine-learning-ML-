```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
x = [2,3,4,5,3,5,6,7]
```

```python
y = [3,6,4,7,4,2,4,6]
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,8))  #width, height
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(8,8))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label= 'test')
ax.plot(y,x,label='test2')
plt.legend()   # it shows colour wise name
plt.show()
```

```python
import seaborn as sns
sns.set_style('whitegrid')
```

```python
tips = sns.load_dataset('tips')
```

```python
type(tips)
```

```python
tips
```

```python
tips['total_bill']
```

```python
sns.distplot(tips['total_bill'])
plt.show()
```

```python
sns.distplot(tips['tip'])
```

```python
sns.jointplot(x='total_bill', y= 'tip', hue= 'smoker' ,data=tips)
```

```python
sns.jointplot(x='total_bill', y= 'tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y= 'tip', data=tips, kind='reg')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.pairplot(tips, hue='sex', palette='ocean') 
```
